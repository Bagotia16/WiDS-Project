Machine Learning

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
data.head()
```

```python
data['Role'] = labelencoder.fit_transform(data['Role'])
```

```python
data.head()
```

```python
x=data.drop(['Player','Role','BBI'],axis=1)
y=data['Role']
```

SVM

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4,random_state=101)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)

predictions1 = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions1))
```

```python
print(confusion_matrix(y_test,predictions1))
```

```python
svm_sc = model.score(x_test, y_test)
svm_sc
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}

grid = GridSearchCV(SVC(),param_grid,verbose=3)
grid.fit(x_train,y_train)
grid_predictions= grid.predict(x_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
grid_sc = grid.score(x_test, y_test)
grid_sc
```

KNN

```python
from sklearn.model_selection import train_test_split
x_train2, x_test2, y_train2, y_test2 = train_test_split(x, y, test_size=0.4,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train2,y_train2)

predictions2 = knn.predict(x_test2)
```

```python
print(confusion_matrix(y_test2,predictions2))
```

```python
print(classification_report(y_test2,predictions2))
```

```python
knn.score(x_test2, y_test2)
```

```python
error_rate=[]

for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate)
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=15)
knn.fit(x_train2,y_train2)

pred = knn.predict(x_test2)
```

```python
print(confusion_matrix(y_test2,pred))
```

```python
print(classification_report(y_test2,pred))
```

```python
KNN_sc = knn.score(x_test2, y_test2)
KNN_sc
```

D-TREE

```python
x_train3,x_test3,y_train3,y_test3=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree= DecisionTreeClassifier()

dtree.fit(x_train3,y_train3)

predictions3 =dtree.predict(x_test3)
```

```python
print(confusion_matrix(y_test3,predictions3))
```

```python
print(classification_report(y_test3,predictions3))
```

```python
dtree_sc = dtree.score(x_test3, y_test3)
dtree_sc
```
